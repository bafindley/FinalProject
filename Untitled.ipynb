```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
file = "csv/train.csv"
```

```python
train_df = pd.read_csv(file)
train_df = train_df.drop(['id', 'belongs_to_collection', 'homepage', 'original_language', 
                          'overview', 'poster_path', 'release_date', 'runtime',
                         'status', 'imdb_id', 'production_countries', 'tagline', 'title', 'Keywords',
                         'crew'], axis=1)
  
train_df = train_df.dropna(how='any')
train_df
```

```python
budpop = pd.read_csv(file)
budpop = budpop[['budget', 'popularity', 'revenue']]
budpop
```

```python


X = budpop[['budget', 'popularity']]
y = budpop['revenue'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Test Score: {testing_score}")
```

```python
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
### END SOLUTION
```
